```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/bank .csv')

object_columns = df.select_dtypes(include='object').columns

unique_values = {col: df[col].unique() for col in object_columns}

null_values = df.isnull().sum()

df_numeric = df.drop(columns=object_columns)

df_numeric.to_csv('banknumericdata.csv', index=False)

df_numeric = pd.read_csv('banknumericdata.csv')
summary_statistics = df_numeric.describe()

print("Object Columns:")
display(pd.DataFrame(object_columns, columns=["Object Columns"]))

print("\nUnique Values of Object Columns:")
unique_values_df = pd.DataFrame.from_dict(unique_values, orient='index').T
display(unique_values_df)

print("\nNull Values in Each Column:")
null_values_df = pd.DataFrame(null_values, columns=["Null Values"])
display(null_values_df)
print("\nSummary Statistics:")
display(summary_statistics)
```

```python
import pandas as pd

df_titanic = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
df_titanic_encoded = pd.get_dummies(df_titanic, columns=['Embarked'], prefix='Embarked')
df_titanic_encoded.head()
```

Problem 4

```python
import seaborn as sns
import matplotlib.pyplot as plt

survival_by_gender = df_titanic_encoded.groupby('Sex')['Survived'].mean()
print(survival_by_gender)

sns.countplot(data=df_titanic_encoded, x='Sex', hue='Survived')
plt.title("Survival Distribution by Gender")
plt.show()
```

Problem 5

```python
sns.countplot(data=df_titanic_encoded, x='Embarked_C', hue='Survived')
sns.countplot(data=df_titanic_encoded, x='Embarked_Q', hue='Survived')
sns.countplot(data=df_titanic_encoded, x='Embarked_S', hue='Survived')
plt.title("Survival Distribution by Port of Embarkation")
plt.show()
```

Problem 6

```python
df_titanic_encoded['Age_group'] = pd.qcut(df_titanic_encoded['Age'], 5, labels=False)

survival_by_class_age = df_titanic_encoded.groupby(['Pclass', 'Age_group'])['Survived'].mean()
survival_table = survival_by_class_age.unstack()

from IPython.display import display, Markdown
display(Markdown(survival_table.to_markdown()))

sns.heatmap(survival_table, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Survival Rate by Age Group and Passenger Class")
plt.show()
```
